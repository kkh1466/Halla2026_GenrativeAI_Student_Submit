```python
# =====================================================================
# [Step 0] D드라이브 캐시 설정
# =====================================================================
import os
import warnings
from transformers.utils import logging

os.environ["TRANSFORMERS_VERBOSITY"] = "error"

# FutureWarning 숨기기
warnings.filterwarnings("ignore", category=FutureWarning)

# Transformers 경고 숨기기
logging.set_verbosity_error()

hf_cache_dir = "D:/HuggingFace_Models"
os.makedirs(hf_cache_dir, exist_ok=True)
os.environ["HF_HOME"] = hf_cache_dir

from transformers import pipeline
from sentence_transformers import SentenceTransformer, util

print("==============================================")
print(" Hugging Face 모델 실습 ")
print("==============================================\n")
```

```python

# =====================================================================
# 1. 감정분석 (Sentiment Analysis)
# =====================================================================
print("🎭 [1. 감정분석]")

sentiment_model = pipeline(
    "text-classification",
    model="dlckdfuf141/korean-emotion-kluebert-v2"
)

# 숫자 라벨 → 감정 이름
emotion_labels = {
    0: "😨 공포",
    1: "😲 놀람",
    2: "😡 분노",
    3: "😢 슬픔",
    4: "😐 중립",
    5: "😊 행복",
    6: "🤢 혐오"
}

texts = [
    "무서운 이야기를 봐서 무섭다...",
    "친구가 약속을 안 지켜서 화가 난다.",
    "갑자기 시험이라니?! 당황스러워.",
    "로맨스 영화를 보면서 너무 슬퍼서 눈물이 났어.",
    "오늘은 남자친구랑 오랜만에 데이트를 하기로 했어!"
]

for t in texts:
    result = sentiment_model(t)

    label_num = result[0]["label"]      # 숫자 라벨
    score = result[0]["score"]          # 신뢰도

    print()
    print("입력 :", t)
    print("감정 :", emotion_labels[label_num])
    print(f"신뢰도 : {score:.4f}")

print(f"\n사용 모델 : {sentiment_model.model.name_or_path}")
```

```python


# =====================================================================
# 2. 번역 (한국어 → 영어)
# =====================================================================
print("🌐 [2. 번역]")

translation_model = pipeline(
    "translation",
    model="Helsinki-NLP/opus-mt-ko-en"
)

kor_text = "오늘은 허깅페이스 실습을 하고 있습니다."

translation = translation_model(kor_text)

print("한국어 :", kor_text)
print("번역 결과 :", translation)
print()

print(f"사용 모델 : {translation_model.model.name_or_path}")
```

```python
# =====================================================================
# 3. 요약
# =====================================================================
print("📝 [3. 요약]")

summarizer = pipeline(
    "summarization",
    model="eenzeenee/t5-base-korean-summarization"
)

long_text = """
인공지능 기술은 최근 다양한 산업에서 활용되고 있다.
특히 자연어 처리 기술은 챗봇, 번역, 문서 요약 등에서 뛰어난 성능을 보여주고 있다.
딥러닝 모델은 많은 데이터를 학습하여 높은 정확도를 달성하지만,
모델의 크기가 커질수록 많은 컴퓨팅 자원이 필요하다.
"""

summary = summarizer(
    long_text,
    max_length=50,
    min_length=10
)

print("원문")
print(long_text)

print("\n요약")
print(summary[0]["summary_text"])


print(f"사용 모델 : {summarizer.model.name_or_path}")
```

```python



# =====================================================================
# 4. 문장 유사도 비교
# =====================================================================
print("👯 [4. 문장 유사도]")

sim_model = SentenceTransformer(
    "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
)

sentence1 = "그는 인공지능 공부를 열심히 한다."
sentence2 = "그는 AI 실습에 집중하고 있다."
sentence3 = "오늘 저녁은 치킨을 먹을 예정이다."

emb1 = sim_model.encode(sentence1)
emb2 = sim_model.encode(sentence2)
emb3 = sim_model.encode(sentence3)

score12 = util.cos_sim(emb1, emb2)
score13 = util.cos_sim(emb1, emb3)

print(sentence1)
print(sentence2)
print(f"유사도 : {score12.item():.4f}")

if score12.item() > 0.7:
    print("→ 두 문장은 매우 유사합니다.")
elif score12.item() > 0.4:
    print("→ 두 문장은 어느 정도 관련이 있습니다.")
else:
    print("→ 두 문장은 관련성이 낮습니다.")

print()

print(sentence1)
print(sentence3)
print(f"유사도 : {score13.item():.4f}")

if score13.item() > 0.7:
    print("→ 두 문장은 매우 유사합니다.")
elif score13.item() > 0.4:
    print("→ 두 문장은 어느 정도 관련이 있습니다.")
else:
    print("→ 두 문장은 관련성이 낮습니다.")

print("\n==============================================")
print("실습 완료!")
print("==============================================")
```
