#3. 요약
- summarization
- text2text-generation

````python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("digit82/kobart-summarization")
model = AutoModelForSeq2SeqLM.from_pretrained("digit82/kobart-summarization")
````

````python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("digit82/kobart-summarization")
model = AutoModelForSeq2SeqLM.from_pretrained("digit82/kobart-summarization")

def summarize(text, max_length=30, min_length=5):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    summary_ids = model.generate(
        **inputs, # Pass input_ids and attention_mask
        num_beams=1, # Changed to greedy search to better control length and repetition
        max_length=max_length, # Further reduced max_length
        min_length=min_length, # Further reduced min_length
        no_repeat_ngram_size=6, # Increased for better summary quality
        early_stopping=True,
        eos_token_id=tokenizer.eos_token_id,
        pad_token_id=tokenizer.pad_token_id,
        decoder_start_token_id=tokenizer.bos_token_id,
        repetition_penalty=10.0, # Greatly increased repetition penalty
        length_penalty=10.0, # Greatly increased length penalty
        do_sample=False # Ensure deterministic greedy search
    )
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)

text = """
인공지능은 최근 다양한 분야에서 활용되고 있다.
의료, 금융, 교육, 자율주행 등에서 많은 성과를 내고 있으며
앞으로도 발전 가능성이 매우 크다.
"""

result = summarize(text, max_length=30, min_length=5)
print(f"Original text: {text}")
print(f"Summary: {result}")
````

#4. 문장 유사도 비교
- sentence-similarity
- sentence-transformers 라이브러리 사용

````python
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")
model = AutoModel.from_pretrained("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")
````

````python
from transformers import AutoTokenizer, AutoModel
import torch
import torch.nn.functional as F

# 모델 및 토크나이저 불러오기
tokenizer = AutoTokenizer.from_pretrained(
    "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
)
model = AutoModel.from_pretrained(
    "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
)

# 비교할 문장
sentence1 = "오늘 날씨가 정말 좋다."
sentence2 = "오늘은 날씨가 맑아서 기분이 좋다."

# 토큰화
inputs = tokenizer(
    [sentence1, sentence2],
    padding=True,
    truncation=True,
    return_tensors="pt"
)

# 임베딩 생성
with torch.no_grad():
    outputs = model(**inputs)

# Mean Pooling
embeddings = outputs.last_hidden_state.mean(dim=1)

# 코사인 유사도 계산
similarity = F.cosine_similarity(
    embeddings[0].unsqueeze(0),
    embeddings[1].unsqueeze(0)
)

print(f"문장 1 : {sentence1}")
print(f"문장 2 : {sentence2}")
print(f"유사도 : {similarity.item():.4f}")
````
